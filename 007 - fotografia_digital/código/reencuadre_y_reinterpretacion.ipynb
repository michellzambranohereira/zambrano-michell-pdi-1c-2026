```python
import cv2
import matplotlib.pyplot as plt
imagen = cv2.cvtColor(cv2.imread("C:\\Users\\usuario\\zambrano-michell-pdi-1c-2026\\007 - fotografia_digital\\imagenes\\originales\\imagen_concierto.jpeg"), cv2.COLOR_BGR2RGB)


# Crop A -> orquesta
recorte_a = imagen[350:950, 300:1320]

# Crop B -> público
recorte_b = imagen[850:1200, 200:1200]

#Dibujar rectángulos en original


imagen_marcada = imagen.copy()

#Rectángulo A (rojo)
cv2.rectangle(imagen_marcada,
              (300, 350),
              (1320, 950),
              (255, 0, 0),
              8)

#Rectángulo B (verde)
cv2.rectangle(imagen_marcada,
              (200, 850),
              (1200, 1200),
              (0, 255, 0),
              8)


#Mostrar resultados

plt.figure(figsize=(16, 10))

#Imagen original
plt.subplot(1, 3, 1)
plt.imshow(imagen_marcada)
plt.title("Imagen original")
plt.axis("off")

#Recorte A
plt.subplot(1, 3, 2)
plt.imshow(recorte_a)
plt.title("Recorte A - Orquesta")
plt.axis("off")

#Recorte B
plt.subplot(1, 3, 3)
plt.imshow(recorte_b)
plt.title("Recorte B - Público")
plt.axis("off")

plt.tight_layout()
plt.show()
```
