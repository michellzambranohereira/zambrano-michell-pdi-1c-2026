4. Fotografía de simplicidad visual

```python
import cv2
import matplotlib.pyplot as plt

#Leer imagen
imagen = cv2.imread("C:\\Users\\usuario\\zambrano-michell-pdi-1c-2026\\007 - fotografia_digital\\imagenes\\originales\\imagen_laguna.jpeg")

#Convertir a RGB
imagen_rgb = cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB)


#Recortar imagen para acercar y eliminar elementos

#Menos grama abajo
#Mucho cielo para mantener espacio negativo

recorte = imagen_rgb[180:1200, 120:780]


#Convertir a escala de grises


gris = cv2.cvtColor(recorte, cv2.COLOR_RGB2GRAY)


#Mostrar imagen original y final


fig, ax = plt.subplots(1, 2, figsize=(14,8))

#Imagen original
ax[0].imshow(imagen_rgb)
ax[0].set_title("Imagen Original")
ax[0].axis("off")

#Imagen final
ax[1].imshow(gris, cmap="gray")
ax[1].set_title("Espacio Negativo + Acercamiento + Escala de Grises")
ax[1].axis("off")

plt.tight_layout()
plt.show()
```

```python
#Guardar imagen en escala de grises

cv2.imwrite(
    r"C:\Users\usuario\zambrano-michell-pdi-1c-2026\007 - fotografia_digital\imagenes\procesadas\laguna_grises.jpeg",
    gris
)
```

## Preguntas guía

¿Qué elementos distraían?
los colores intensos del cielo y el agua son los elementos que más re robaban la atención del objeto, que eset caso es la escultura del hombre.

¿Qué se eliminó?
Se realizó un recorte para tratar de reducir estos elementos y reforzar el espacio negativo dentro de la imagen.

¿Cómo cambia la lectura en escala de grises?
La escala de grises simplifica la imagen y dirige la atención hacia las formas, el horizonte y la distribución del espacio vacío.
