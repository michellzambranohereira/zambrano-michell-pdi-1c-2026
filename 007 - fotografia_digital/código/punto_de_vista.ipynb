```python
import cv2
import matplotlib.pyplot as plt

#Leer imágenes
img_lejos = cv2.imread("C:\\Users\\usuario\\zambrano-michell-pdi-1c-2026\\007 - fotografia_digital\\imagenes\\originales\\imagen_lejos.jpeg")
img_cerca = cv2.imread("C:\\Users\\usuario\\zambrano-michell-pdi-1c-2026\\007 - fotografia_digital\\imagenes\\originales\\imagen_cerca.jpeg")

#Convertir BGR a RGB
img_lejos = cv2.cvtColor(img_lejos, cv2.COLOR_BGR2RGB)
img_cerca = cv2.cvtColor(img_cerca, cv2.COLOR_BGR2RGB)

#Crear figura
fig, ax = plt.subplots(1, 2, figsize=(12,8))

#Imagen desde lejos
ax[0].imshow(img_lejos)
ax[0].set_title("Imagen desde lejos")
ax[0].axis("off")

# Imagen desde cerca
ax[1].imshow(img_cerca)
ax[1].set_title("Imagen desde cerca")
ax[1].axis("off")

# Mostrar dentro del notebook
plt.tight_layout()
plt.show()
```
